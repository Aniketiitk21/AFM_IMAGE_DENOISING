```python
#Contour Detection

import cv2
import numpy as np
from google.colab import files
from google.colab.patches import cv2_imshow

from matplotlib import pyplot as plt

uploaded = files.upload()
image_filename = list(uploaded.keys())[0]

image = cv2.imread(image_filename)

image= cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

def apply_canny_edge_detection(image, low_threshold, high_threshold):


    # Apply Gaussian blur to the image to reduce noise
    blurred_image = cv2.GaussianBlur(image, (5, 5), 0)

    # Apply Canny edge detection
    edges = cv2.Canny(image, low_threshold, high_threshold)

    # Display the original and edge-detected images
    plt.figure(figsize=(10, 5))

    plt.subplot(1, 2, 1)
    plt.imshow(image, cmap='gray')
    plt.title('Original Image')

    plt.subplot(1, 2, 2)
    plt.imshow(edges, cmap='gray')
    plt.title('Canny Edge Detection')

    # Find contours in the edge-detected image
    contours, _ = cv2.findContours(edges, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

    hexagon_count = 0
    oblique_count = 0

    # Analyze contours to recognize shapes
    for contour in contours:
        # Approximate the contour to reduce the number of vertices
        epsilon = 0.04 * cv2.arcLength(contour, True)
        approx = cv2.approxPolyDP(contour, epsilon, True)

        # Get the number of vertices
        num_edges = len(approx)

        # Classify the shape based on the number of edges
        if num_edges == 6:
          shape = "Hexagon"
          hexagon_count += 1

        else:
          shape = "Oblique"
          oblique_count += 1

        # Draw the contour on the original image
        cv2.drawContours(image, [approx], 0, (0, 255, 0), 2)

        # # Display the classification result
        # font = cv2.FONT_HERSHEY_SIMPLEX
        # cv2.putText(image, shape, (approx[0][0][0], approx[0][0][1] - 10),
        #             font, 0.5, (255, 0, 0), 1, cv2.LINE_AA)

    # Display the original image with contours and classifications
    plt.imshow(image, cmap='gray')
    plt.title('Recognized Shapes')

    plt.show()

    print("Number of Hexagon Islands: ", hexagon_count)
    print("Number of Oblique Islands: ", oblique_count)

# Specify the input image path and Canny edge detection thresholds
low_threshold = 180
high_threshold = 300

# Apply Canny edge detection
apply_canny_edge_detection(image, low_threshold, high_threshold)
```

```python
#Image Descriptor Hu Moments

import cv2
import numpy as np
from google.colab import files
from google.colab.patches import cv2_imshow

from matplotlib import pyplot as plt

uploaded = files.upload()
image_filename = list(uploaded.keys())[0]

image = cv2.imread(image_filename)

image= cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

def compute_hu_moments(image, low_threshold, high_threshold):


    # Apply Gaussian blur to the image to reduce noise
    blurred_image = cv2.GaussianBlur(image, (5, 5), 0)

    # Apply Canny edge detection
    edges = cv2.Canny(image, low_threshold, high_threshold)

    # Display the original and edge-detected images
    plt.figure(figsize=(10, 5))

    plt.subplot(1, 2, 1)
    plt.imshow(image, cmap='gray')
    plt.title('Original Image')

    plt.subplot(1, 2, 2)
    plt.imshow(edges, cmap='gray')
    plt.title('Canny Edge Detection')

    # Find contours in the edge-detected image
    contours, _ = cv2.findContours(edges, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

    # Initialize a list to store Hu Moments for each contour
    hu_moments_list = []

    # Analyze contours to compute Hu Moments
    for contour in contours:
        # Approximate the contour to reduce the number of vertices
        epsilon = 0.04 * cv2.arcLength(contour, True)
        approx = cv2.approxPolyDP(contour, epsilon, True)

        # Get the Hu Moments
        moments = cv2.moments(approx)
        hu_moments = cv2.HuMoments(moments).flatten()

        # Store the Hu Moments in the list
        hu_moments_list.append(hu_moments)

    return hu_moments_list

# Specify the input image path and Canny edge detection thresholds
low_threshold = 180
high_threshold = 300

# Compute Hu Moments for each contour
hu_moments_list = compute_hu_moments(image, low_threshold, high_threshold)

# Display the computed Hu Moments
for i, hu_moments in enumerate(hu_moments_list):
    print(f'Contour {i + 1} - Hu Moments: {hu_moments}')
```

```python
# Import the necessary packages
import numpy as np
import matplotlib.pyplot as plt
from skimage import io, color, util, restoration
from sklearn.feature_extraction import image
from sklearn.decomposition import MiniBatchDictionaryLearning

# Load the noisy image
# Replace with the path to your image file
image_path = '/content/1_amplitude.jpg'
noisy_image = io.imread(image_path)
noisy_image = util.img_as_float(noisy_image)
print('Noise Image shape:',noisy_image.shape)

# Extract small patches from the grayscale image
patch_size = (2, 2)
patches = image.extract_patches_2d(noisy_image, patch_size)
print('Number of Patches :',patches.shape[0])
print('Shape of patches:',patches.shape)
# Reshape the patches for dictionary learning
data = patches.reshape(patches.shape[0], -1)
print('Shape of Input data :',data.shape)

# Perform dictionary learning
n_components = 1000 # Number of dictionary atoms to learn
dl = MiniBatchDictionaryLearning(n_components=n_components,
								alpha=1.0,
								n_iter=500)
# training
dl.fit(data)

# Denoise the patches using the learned dictionary
denoised_patches = np.dot(dl.transform(data), dl.components_)
print('Shape of Output Denoised patches:',denoised_patches.shape)

# Reshape the denoised patches back to their original shape
denoised_patches = denoised_patches.reshape(patches.shape)
print('After Reshaping, Output Denoised patches:',denoised_patches.shape)

# Reconstruct the denoised image from the patches
reconstructed_image = image.reconstruct_from_patches_2d(denoised_patches, noisy_image.shape)
print('reconstructed_image:',reconstructed_image.shape)

# Show the original noisy image and the denoised image
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(noisy_image)
plt.title('Noisy Image')
plt.axis('off')
plt.subplot(1, 2, 2)
plt.imshow(reconstructed_image)
plt.title('Denoised Image')
plt.axis('off')
plt.show()
```
